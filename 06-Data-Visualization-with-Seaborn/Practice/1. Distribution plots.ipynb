# Seaborn is data visualization library built on top of matplotlib and it is used to plot statistical graphs

# Distribution plots

```python
import seaborn as sns
```

```python
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
type(tips)
```

```python
# it shows distribution of univariate variable means only one variable
sns.distplot(tips['total_bill'])
```

```python
sns.distplot(tips['total_bill'],kde=False,bins=40)
```

```python
# joinplot is used combine two dist plots of bivariate data  - two variable
sns.jointplot(x='total_bill',y='tip',data=tips,kind='hex')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='reg')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='kde')
```

```python
# pairplots is going to plot pair vise relationship across the entire dataset atleast for numerical columns
# hue argument is used to plot categorical column
# palette allows us to color this with specific color
sns.pairplot(tips,hue='sex',palette='coolwarm')
```

```python
# rugplots draws dash for each point
sns.rugplot(tips['total_bill'])
```

```python
# kde kernal density  estimation is sum of all the noraml dist around rugplot (Gauss dist)
sns.distplot(tips['total_bill'])
```

```python
# only kde plot without histogram
sns.kdeplot(tips['total_bill'])
```
