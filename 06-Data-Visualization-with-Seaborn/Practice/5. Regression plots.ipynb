```python
import seaborn as sns
```

```python
%matplotlib inline
tips = sns.load_dataset('tips')
tips.head()
```

```python
# lmplots allows us to display linear model with seaborn
# it is calling or using matplotlib underhood
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',markers=['o','*'],
          scatter_kws={'s':100})
# we can also hue for categorical column
```

```python
# hue separates by column 
# col create separate graph for each value
sns.lmplot(x='total_bill',y='tip',data=tips,col='sex')
```

```python
# hue separates by column 
# col create separate graph for each value
# row create separate graph for each value in row
sns.lmplot(x='total_bill',y='tip',data=tips,col='sex',row='time')
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='day',hue='sex',
          aspect=0.6,height=8)
```
