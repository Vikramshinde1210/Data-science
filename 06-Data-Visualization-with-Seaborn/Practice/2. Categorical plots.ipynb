```python
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
# bar plot - visualization of group by action
# it showing avg total bill
sns.barplot(x='sex',y='total_bill',data=tips)
```

```python
import numpy as np
```

```python
# default estimator is avg function
sns.barplot(x='sex',y='total_bill',data=tips,estimator=np.std)
```

```python
# countplot it is similar to barplot instead here estimator is explicitely counting no of occurences
# same as tips.count() using pandas
sns.countplot(x='sex',data=tips)
```

```python
# boxplots used to show dist of categorical data
sns.boxplot(x='day',y='total_bill',data=tips,hue='sex')
```

```python
# it is similar to the boxplot allows us to plot kde of underline data
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex')
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex',split=True)
```

```python
# strip plot draw scatter plot where one variable uis categorical
# this is just a scatter plot based on the category
# jitter is used to separate stacked points
sns.stripplot(x='day',y='total_bill',data=tips,jitter=True,hue='sex',split=True)
```

```python
# combining striplot with violin plot = warm plot
# don't use swarm plots for very large datasets
```

```python
sns.swarmplot(x='day',y='total_bill',data=tips)
```

```python
# comnining swarm with violin plot
sns.violinplot(x='day',y='total_bill',data=tips)
sns.swarmplot(x='day',y='total_bill',data=tips,color='black')
```

```python
# factor plot
# more general method to plot the all graphs using kind argument
sns.factorplot(x='day',y='total_bill',data=tips,kind='violin')
```
