# Matplotlib

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0,5,11)
y = x ** 2
```

```python
x
```

```python
y
```

# Fuctional method

```python
plt.plot(x,y)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
# plt.show() if you're not using notebook
```

# Creating multi plots on same canvas

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')
plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

# Object oriented method

```python
fig = plt.figure()   # creates blank canvas

axes = fig.add_axes([0.1,0.1,0.8,0.8])

axes.plot(x,y)
axes.set_xlabel('X label')
axes.set_ylabel('Y label')
axes.set_title('Title')
```

```python
# putting two sets of figure on one canvas
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])   # meaning of 0.1 means 10% from left 

axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y)
axes1.plot('Larger plot')

axes2.plot(y,x)
axes2.set_title('Smaller plot')
```

# Creating subplots using OOP


```python
x = [1,2,3,4,5,6,7]
y = [i**2 for i in x]
```

```python
fig,axes = plt.subplots(1,2) # rows ,columns

# ploting graph by iterating
# for cr in axes:
#     cr.plot(x,y)


# ploting graph using index
axes[0].plot(x,y)
axes[0].set_title('First')
axes[1].plot(y,x)
axes[1].set_title('Second')

# to avoid overlapping issues 
# always put at the end
plt.tight_layout()
```


# Figure size and DPI

```python
# figsize is tuple of width and height in inches
fig  = plt.figure(figsize=(8,2),dpi=100)
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig,axes  = plt.subplots(2,1,figsize=(8,2),dpi=100)

axes[0].plot(x,y)
axes[1].plot(y,x)

plt.tight_layout()
```

# Legends

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,label='X vs Y')
ax.plot(y,x, label='Y vs X')
# ax.legend(loc=10)
ax.legend(loc=(0.1,0.1))
```

# Styling graph

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
# ax.plot(x,y,color='orange')
ax.plot(x,y,color='#FF8C00',lw=3,alpha=0.5)
# color name ,rgb color both accepted
# alpha - how transparent the line is
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='#FF8C00',lw=3,linestyle='--',marker='o',markersize=20,
       markerfacecolor='yellow',markeredgecolor='black',markeredgewidth=3)

# values for linestyle = steps, -- ,: , _. , -
# ls shortform for linestyle
```

# Control over axis appearance

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='#FF8C00',lw=3,linestyle='--')

# pass list of lower bound and upper bound 
ax.set_xlim([0,2])
ax.set_ylim([0,2])
```
