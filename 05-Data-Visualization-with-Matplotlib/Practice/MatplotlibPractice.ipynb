```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x=np.arange(0,11)
```

```python
x
```

```python
x.size
```

```python
y=x**2
y
```

```python
# plotting the graph using the x and y co-ordinates
plt.plot(x,y,'g--')
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("My first plot")
plt.show()
```

```python
# create multiplots and same canvas
plt.plot(x,y,'r--')
plt.plot(y,x,'g--')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')
plt.subplot(1,2,2)
plt.plot(y,x,'g')
```

```python
fig=plt.figure()
axes=fig.add_axes([0.1,0.1,0.8,0.8]) # distance from left,bottom,width and height
axes.plot(x,y,'r')
axes.set_xlabel('X-axis')
axes.set_ylabel('Y-axis')
axes.set_title("figure")
```

```python
fig=plt.figure()
axes1=fig.add_axes([0.1,0.1,0.8,0.8])
axes2=fig.add_axes([0.2,0.5,0.4,0.3])
axes1.plot(x,y,'r')
axes2.plot(y,x,'g')
```

```python
fig,ax=plt.subplots(figsize=(10,5))
# - means the solid line
# -- means the dashed line
# -. 
# : complete dotted line
ax.plot(x,x+1,color="purple",lw=3,ls='-',marker='s',markersize=10,markerfacecolor='yellow',markeredgecolor="green",markeredgewidth=5) 
# o is used for solid circular markers
# + 
# s is used for solid square 
```

```python
# alpha=0 not visible

# used for opacity /tranparency 
# 1 completely visible
#for color we can mentioned rgb or hex color codes also
```

```python
fig,ax=plt.subplots(1,2,figsize=(12,4))
ax[0].plot(x,x**2,x,x**3)
ax[0].axis('tight')
ax[1].plot(x,x**2,x,x**3)
ax[1].set_xlim([4,6])
ax[1].set_ylim([0,400])
```
