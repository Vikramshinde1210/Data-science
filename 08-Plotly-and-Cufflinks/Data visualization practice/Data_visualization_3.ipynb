```python

import matplotlib.pyplot as plt
# it contains 83 module
import numpy as np
%matplotlib inline
# two methods of drawing graph 
# %matplotlib is used to open the plot embedded in the notebook or any interactive shell 
# mention this line at the beginning of the program
# if you run file separately use plt.show() to display graph in new window
# it is mentioned at the end of the program
```

```python
x=np.array([i for i in range(0,11)])
x
```

```python
y= x**2
y
```

```python
plt.plot(x,y,'r--')
# mention third parameter as color's first letter
# use -- after color's first letter for styling 
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('First plot')
```

```python
# create multiplots on same canvas
```

```python
plt.plot(x,y,'r')
plt.plot(y,x,'g')
```

```python
# plt.subplots(r,c,n)
# here n [1,2,3.....] start from upper left corner and increases towards right
# n is number of canvas which you want to select
```

```python
# first graph
plt.subplot(1,2,1)
plt.plot(x,y,'r')
# second graph on same canvas
plt.subplot(1,2,2)
plt.plot(y,x,'g')
```

```python
# using object oriented methods of figure object
fig=plt.figure()
axes=fig.add_axes([0.1,0.1,0.8,0.8]) # distance from left,bottom,width and height
axes.plot(x,y,'r')
axes.set_xlabel('X-axis')
axes.set_ylabel('Y-axis')
axes.set_title("figure")
```

```python
# this two plots will occur in the same box
fig1= plt.figure()
axes1= fig1.add_axes([0.1,0.1,0.8,0.8])
axes1.set_xlabel('X')
axes2= fig1.add_axes([0.2,0.5,0.4,0.3])
axes2.set_xlabel('X1')
axes1.plot(x,y,'r')
axes2.plot(y,x,'g')
```

```python
# styling of plots or curves
```

```python
fig ,axes=plt.subplots(figsize=(10,5))
axes.plot(x,x+1,'r--')
# -- for dashed line
# 
```

```python
fig ,axes=plt.subplots(figsize=(10,5))
axes.plot(x,x+1,color='red')
# color attribute
```

```python

fig ,axes=plt.subplots(figsize=(10,5))
axes.plot(x,x+1,color='purple',lw=3,ls='-')
# to set the line width use linewidth paramater you can also use lw
# default lw is 1
# linestyle or ls to change the style
# if mention '- ' solid line, '--' dashed line, '-.' dash dot line,
# ':' complete dotted line
```

```python
fig ,axes=plt.subplots(figsize=(10,5))
axes.plot(x,x+1,color='purple',lw=3,ls='-',marker='o')
# marker is the explicit point which you plot on the graph
# 'o' is used for circular
# '+' used for + sign
# 's' used for solid square
# '1' used for vertical lines
# '2' used for slanted lines
# '*' for stars
```

```python
fig ,axes=plt.subplots(figsize=(10,5))
axes.plot(x,x+1,color='purple',lw=3,ls='-',marker='o',markersize=5)
# use markersize attribute to change the size of the marker
```

```python
fig ,axes=plt.subplots(figsize=(10,5))
axes.plot(x,x+1,color='purple',lw=3,ls='-',marker='o',markersize=10,markerfacecolor='y',markeredgecolor='green',markeredgewidth=5)
#markerfacecolor for marker color boundry remains same
# if you want to change boundry color use markeredgecolor attribute
# width of boundry markeredgewidth
```

```python
fig ,ax= plt.subplots(figsize=(10,5))
ax.plot(x,x+1,color='purple',lw=3,alpha=1)
# alpha is used to show the opacity
# 0 not visible complete tranparent
# 1 complete visible
```

```python
# set limit to your axes
fig ,axes=plt.subplots(1,2,figsize=(12,4))
axes[0].plot(x,x**2,x,x**3)
axes[0].axis('tight')
# axis tight is used to show the entire graph
axes[1].plot(x,x**2,x,x**3)
axes[1].set_xlim([4,6])
axes[1].set_ylim([0,400])
# if you want to see the graph between specific range only use set_xlim method
```
